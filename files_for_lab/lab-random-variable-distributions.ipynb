```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
```

1. Use the numerical variables from our dataset this week.

```python
numerical_df = pd.read_csv('numerical.csv')
numerical_df.head()
```

2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for column in numerical_df.columns:
    
    plt.figure(figsize=(8, 4))
    sns.histplot(numerical_df[column], kde=False, bins=30)
    plt.title(f'Distribution Plot for {column}')
    
    if expon.fit(numerical_df[column])[1] > 0.05:
        print(f"{column} follows an exponential distribution.")
    elif norm.fit(numerical_df[column])[1] > 0.05:
        print(f"{column} follows a normal distribution.")
    else:
        print(f"{column} does not follow a normal or exponential distribution.")
    
    plt.show()
```

3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
import seaborn as sns


```python
for numerical_variable in numerical_df.columns:
    if numerical_variable != 'total_claim_amount':
        fig, axes = plt.subplots(1, 3, figsize=(16, 4))
        fig.suptitle(numerical_variable)

        # Original data:
        sns.histplot(numerical_df[numerical_variable], kde=True, ax=axes[1])
        axes[1].set_title('Original Data')
        
        # Check if the original data follows a normal or exponential distribution
        if expon.fit(numerical_df[numerical_variable])[1] > 0.05:
            print("The Original follows an exponential distribution.")
        elif norm.fit(numerical_df[numerical_variable])[1] > 0.05:
            print("The Original follows a normal distribution.")
        else:
            print("The Original does not follow a normal or exponential distribution.")

        # Log transformation:
        log_transformed = np.log1p(numerical_df[numerical_variable])
        sns.histplot(log_transformed, kde=True, ax=axes[0])
        axes[0].set_title('Log-Transformed Data')
        
        # Check if the log-transformed data follows a normal or exponential distribution
        if expon.fit(log_transformed)[1] > 0.05:
            print(f"Log-transformed {numerical_variable} follows an exponential distribution.")
        elif norm.fit(log_transformed)[1] > 0.05:
            print(f"Log-transformed {numerical_variable} follows a normal distribution.")
        else:
            print(f"Log-transformed {numerical_variable} does not follow a normal or exponential distribution.")

        # Square root transformation:
        sqrt_transformed = np.sqrt(numerical_df[numerical_variable])
        sns.histplot(sqrt_transformed, kde=True, ax=axes[2])
        axes[2].set_title('Square Root-Transformed Data')

        # Check if the square root-transformed data follows a normal or exponential distribution
        if expon.fit(sqrt_transformed)[1] > 0.05:
            print(f"Square root-transformed {numerical_variable} follows an exponential distribution.")
        elif norm.fit(sqrt_transformed)[1] > 0.05:
            print(f"Square root-transformed {numerical_variable} follows a normal distribution.")
        else:
            print(f"Square root-transformed {numerical_variable} does not follow a normal or exponential distribution.")

        plt.show()
```
